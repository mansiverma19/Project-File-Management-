# Project File Management

```python
import pandas as pd
import csv
from colorama import Fore
from tabulate import tabulate as tb
```

```python
#reads the file
def open_file():
    with open('file.csv') as file_obj:
        file=csv.DictReader(file_obj,skipinitialspace=True)
        return list(file)


#returns value
def return_val(y,string,nm=None):
    if y:
        return nm
    else:
        return string

    
#gives project name
def find_project_name(idval):
    file=open_file()
    y=False
    for i in file:
        if i['id']==idval:
            project=i['project name']
            y=True
    return return_val(y,"No Such Id Value",project)
  

#gives project id
def find_project_id(project_nm):
    file=open_file()
    y=False
    for i in file:
        if i['project name'].lower()==project_nm.lower():
            name=i['id']
            y=True
    return return_val(y,"No Such Project Name",nm)


#gives all students who are assigned in a project 
def find_students(idval):
    file=open_file()
    l,y=[],False
    for i in file:
        if i['id']==idval:
            val=list(i.keys())
            y=True
            for j in val:
                if i[j]=='1':
                    l.append(j)
            break
    return return_val(y,"No Such Id Value",l)
        

#give all the projects to which particular person is assigned
def find_projects(st_nm):
    file=open_file()
    l,y=[],False
    for i in file:
        key=list(i.keys())
        key=[j.lower() for j in key]
        val=list(i.values())
        if st_nm in key:
            idx=key.index(st_nm)
            y=True
            if val[idx]=='1':
                l.append(val[1])
    return return_val(y,"No Such Student Name",l)


#give all projects and their ids
def project_id():
    file=open_file()
    d={}
    for i in file:
        d[i['id']]=i['project name']
    return d


#display all details
def all_details():
    file=open_file()
    l=file[0].keys()
    rows= [x.values() for x in file]
    df1=tb(rows,headers=l,tablefmt='pretty')
    print(Fore.BLACK+df1)

    
#adding new project
def write_project():
    df=pd.read_csv('file.csv')
    column=df.columns
    print("--------------------------------------------------------------------------------------------")
    print("TYPE '1' IF YOU WANT TO ASSIGN THE PROJECT ELSE TYPE '0' IN STRING FORMAT EXAMPLE=01100101")
    id_=input("ENTER THE PROJECT ID : ")
    project_nm=input("ENTER THE PROJECT NAME : ")
    l=[id_,project_nm]
    print("NAME Of STUDENTS")
    i=2
    while(i<len(column)):
        n=input(column[i]+':')
        if n=='0' or n=='1':
            l.append(n)
            i+=1
        else:
            print("Wrong input")
    df.loc[len(df.index)]=l
    df.to_csv('file.csv',index=False)
    

#adding new student
def write_student():
    df=pd.read_csv('file.csv')
    p_nm=list(df['project name'])
    print(p_nm,type(p_nm))
    print("--------------------------------------------------------------------------------------------")
    print("TYPE '1' IF YOU WANT TO ASSIGN THE PROJECT ELSE TYPE '0' IN STRING FORMAT EXAMPLE=01100101")
    student_nm=input("ENTER THE STUDENT NAME : ")
    print("NAME Of STUDENTS")
    i,l=0,[]
    while(i<len(p_nm)):
        n=input(p_nm[i]+':')
        if n=='0' or n=='1':
            l.append(n)
            i+=1
        else:
            print("Wrong input")
    df[student_nm]=l
    df.to_csv('file.csv',index=False)
```

```python
print(Fore.RED+"----------------------------------------------------------------------")
print(Fore.BLUE+"1 : Display title of a Project Title when user enters the Project ID")
print(Fore.BLUE+"2 : Display Project ID of a project when user enters the Project Title")
print(Fore.BLUE+"3 : Display students working on a project with given Project ID")
print(Fore.BLUE+"4 : Display the list of all Projects on which an student is working")
print(Fore.BLUE+"5 : Display All Projects and their IDs")
print(Fore.BLUE+"6 : Display all Details stored in the text file in a tabular format")
print(Fore.BLUE+"7 : Insert new Project")
print(Fore.BLUE+"8 : Insert new Student")
print(Fore.RED+"----------------------------------------------------------------------")
print(Fore.BLACK)
for i in range(8):
    print(Fore.RED+"ENTER YOUR CHOICE : ",end="")
    ch=int(input())
    if ch==1:
        print(Fore.BLUE+"ENTER PROJECT ID : ",end="")
        idval=input()
        nm=find_project_name(idval)
        print(Fore.GREEN+"NAME OF PROJECT : ",nm)
    elif ch==2:
        print(Fore.BLUE+"ENTER PROJECT NAME : ",end="")
        project_nm=input()
        id_=find_project_id(project_nm)
        print(Fore.GREEN+"ID OF PROJECT : ",id_)
    elif ch==3:
        print(Fore.BLUE+"ENTER PROJECT ID : ",end="")
        idval=input()
        l=find_students(idval)
        print(Fore.GREEN+"students working on a project with given Project ID : ",l)
    elif ch==4:
        print(Fore.BLUE+"ENTER STUDENT NAME : ",end="")
        st_nm=input()
        l=find_projects(st_nm.lower())
        print(Fore.GREEN+"list of all Projects on which an student is working : ",l)
    elif ch==5:
        d=project_id()
        print(Fore.GREEN+"All project and ids : ",d)
    elif ch==6:
        print("TABLE")
        all_details()
    elif ch==7:
        write_project()
        all_details()
    elif ch==8:
        write_student()
        all_details()
    else:
        print(Fore.RED+"INVALID CHOICE !!!")
```
